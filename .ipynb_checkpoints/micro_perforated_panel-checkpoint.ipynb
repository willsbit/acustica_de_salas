### 6. Placa microperfurada

```python
import numpy as np
import matplotlib.pyplot as plt
from macroscopic_models import delany_bazley, jca

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
# define frequency space
f_start = 20
f_end = 10000
f = np.arange(f_start, f_end, step=1)
omega = 2 * np.pi * f

# air properties
rho0 = 1.21  # [kg/m³] air density
c0 = 343  # [m/s] sound speed
k0 = omega / c0 # [rad / m] wave number
air_viscosity = 1.84e-5
```

#### 6.1. Variação da razão de área perfurada

```python
# microperforated panel properties
a = 0.2e-3 # [m] aperture radius
l = 0.4e-3 # [m] thickness of the perforated panel
D = 0.05 # [m] total absorber size
perforated_area_ratios = np.array([0.01, 0.05, 0.2])
boundary_layer_ratio = 2*a*np.sqrt(omega*rho0 / (4*air_viscosity))

# plot setup
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title(r"Efeito da variação da razão de área perfurada no coeficiente de absorção")
ax.set_xlabel("Frequência [Hz]")
ax.set_ylabel(r"$\alpha$ [-]")
ax.set_xlim(f_start, f_end)
ax.set_ylim(-0.2, 1.2)
ax.grid(axis="both", which="both", linestyle="--", alpha=0.4)
labels = [r"$\psi = 1~\text{[%]}$", r"$\psi = 5~\text{[%]}$", r"$\psi = 20~\text{[%]}$"]

for (perforated_area_ratio, label) in zip(perforated_area_ratios, labels):
    R_1 = ((32*air_viscosity*l) / (perforated_area_ratio * (2*a)**2))
    R_2 = np.sqrt(1 + (boundary_layer_ratio**2 / 32)) + ((np.sqrt(2)/32) * boundary_layer_ratio*(2*a / l))
    acoustic_resistance = R_1 * R_2
    panel_surface_density = (rho0*l / perforated_area_ratio) * (1 + np.sqrt(9 + (boundary_layer_ratio**2 / 2)) + 0.85*(2*a/l))
    Zs = acoustic_resistance + 1j*omega*panel_surface_density - ((1j*rho0*c0) / np.tan(k0*D))

    # reflection coefficient
    Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))
    # absorption coefficient
    alpha = 1 - (abs(Vp)) ** 2 
    # plot
    ax.semilogx(f, alpha, label=label)

ax.legend()
plt.show()
```

#### 6.2. Variação do diâmetro dos furos

```python
# microperforated panel properties
aperture_sizes = np.array([0.1e-3, 0.25e-3, 0.5e-3]) # [m] aperture radius
l = 0.4e-3 # [m] thickness of the perforated panel
D = 0.05 # [m] total absorber size
perforated_area_ratio = 0.01

# plot setup
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title(r"Efeito da variação do diâmetro dos furos no coeficiente de absorção")
ax.set_xlabel("Frequência [Hz]")
ax.set_ylabel(r"$\alpha$ [-]")
ax.set_xlim(f_start, f_end)
ax.set_ylim(-0.2, 1.2)
ax.grid(axis="both", which="both", linestyle="--", alpha=0.4)
labels = [r"$2a = 0.2~\text{[mm]}$", r"$2a = 0.5~\text{[mm]}$", r"$2a = 1.0~\text{[mm]}$"]

for (a, label) in zip(aperture_sizes, labels):
    boundary_layer_ratio = 2*a*np.sqrt(omega*rho0 / (4*air_viscosity))
    R_1 = ((32*air_viscosity*l) / (perforated_area_ratio * (2*a)**2))
    R_2 = np.sqrt(1 + (boundary_layer_ratio**2 / 32)) + ((np.sqrt(2)/32) * boundary_layer_ratio*(2*a / l))
    acoustic_resistance = R_1 * R_2
    panel_surface_density = (rho0*l / perforated_area_ratio) * (1 + np.sqrt(9 + (boundary_layer_ratio**2 / 2)) + 0.85*(2*a/l))
    Zs = acoustic_resistance + 1j*omega*panel_surface_density - ((1j*rho0*c0) / np.tan(k0*D))

    # reflection coefficient
    Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))
    # absorption coefficient
    alpha = 1 - (abs(Vp)) ** 2 
    # plot
    ax.semilogx(f, alpha, label=label)

ax.legend()
plt.show()
```

#### 6.3. MPP vs DLMPP

```python
# microperforated panel properties
a = 0.2e-3 # [m] aperture radius
l = 0.4e-3 # [m] thickness of the perforated panel
D = 0.05 # [m] total absorber size
perforated_area_ratio = 0.01
boundary_layer_ratio = 2*a*np.sqrt(omega*rho0 / (4*air_viscosity))

# dlmpp panel properties
mpp_D = 0.025

# plot setup
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title(r"MPP vs DLMPP")
ax.set_xlabel("Frequência [Hz]")
ax.set_ylabel(r"$\alpha$ [-]")
ax.set_xlim(f_start, f_end)
ax.set_ylim(-0.2, 1.2)
ax.grid(axis="both", which="both", linestyle="--", alpha=0.4)
labels = [r"$\psi = 1~\text{[%]}$", r"$\psi = 5~\text{[%]}$", r"$\psi = 20~\text{[%]}$"]

# MPP
R_1 = ((32*air_viscosity*l) / (perforated_area_ratio * (2*a)**2))
R_2 = np.sqrt(1 + (boundary_layer_ratio**2 / 32)) + ((np.sqrt(2)/32) * boundary_layer_ratio*(2*a / l))
acoustic_resistance = R_1 * R_2
panel_surface_density = (rho0*l / perforated_area_ratio) * (1 + np.sqrt(9 + (boundary_layer_ratio**2 / 2)) + 0.85*(2*a/l))
Zs = acoustic_resistance + 1j*omega*panel_surface_density - ((1j*rho0*c0) / np.tan(k0*D))

# reflection coefficient
Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))
# absorption coefficient
alpha = 1 - (abs(Vp)) ** 2 
# plot
ax.semilogx(f, alpha, label="MPP")


# DLMPP
first_layer_impedance = -1j*rho0*c0 / np.tan(k0*mpp_D)
second_layer_impedance = acoustic_resistance + 1j*omega*panel_surface_density - 1j*rho0*c0 / np.tan(k0*mpp_D)

Zs = acoustic_resistance + 1j*omega*panel_surface_density + ((1/first_layer_impedance) + (1/second_layer_impedance))**(-1)
# reflection coefficient
Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))
# absorption coefficient
alpha = 1 - (abs(Vp)) ** 2 
# plot
ax.semilogx(f, alpha, label="DLMPP")


ax.legend()
plt.show()
```
