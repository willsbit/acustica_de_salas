```python
import numpy as np
import matplotlib.pyplot as plt
from macroscopic_models import jca

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.rcParams["savefig.pad_inches"] = 0
```

```python
# define frequency space
f_start = 20
f_end = 10000
f = np.arange(f_start, f_end, step=1)
omega = 2 * np.pi * f

# air properties
rho0 = 1.21  # [kg/m³] air density
c0 = 343  # [m/s] sound speed
k0 = omega / c0 # [rad / m] wave number

# porous material JCA properties
sigma = 15000 # [N*s/m^4] flux resistivity
tortuosity=1.01
porosity=0.98
viscous_characteristic_length=130e-6 # [m]
d = 0.05
```

```python
# porous absorber
Zc, kc = jca(f, sigma=sigma, tortuosity=tortuosity, porosity=porosity, viscous_characteristic_length=viscous_characteristic_length) # JCA parameters
Zs = -1j * (Zc / np.tan(kc * d))  # surface impedance
Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))  # reflection coefficient
alpha_porous = 1 - (abs(Vp)) ** 2  # absorption coefficient
```

```python
# perforated panel
a = 2.5e-3 # [m] aperture radius
l = 3e-3 # [m] thickness of the perforated panel
D = 0.15 # [m] total absorber size
l_corr = l + 1.7*a # [m] corrected lenght of the tube
perforated_area_ratio = 0.15
panel_surface_density = rho0*l_corr / perforated_area_ratios

# porous material properties
d = 0.08 # [m] thickness of porous layer
# porous material characteristic impedance and characteristic wave number
Zp, kp = jca(f, sigma=sigma, tortuosity=tortuosity, porosity=porosity, viscous_characteristic_length=viscous_characteristic_length)

# ressonance frequency
f_res = (c0 / 2*np.pi) * np.sqrt(perforated_area_ratio / (D * l_corr))

# panel impedance
Zpp = 1j*omega*panel_surface_density

# impedance on top of the air layer
Zs_ar = (-1j*rho0*c0) / np.tan(k0*(D - d))

# impedance on top of the porous layer
Zsi = (((-1j*Zs_ar*Zp) / np.tan(kp*d)) + (Zp**2)) / (Zs_ar - (1j*Zp / np.tan(kp*d)))

# total impedance
Zs = Zpp + Zsi

# reflection coefficient
Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))

# absorption coefficient
alpha_perforated = 1 - (abs(Vp)) ** 2
```

```python
# membrane 1 properties
D = 0.25  # [m] size of the air gap
membrane_thickness = 3e-3  # [m] membrane thickness
rho_m = 485  # [kg/m^3] specific mass of the membrane
surface_density = membrane_thickness * rho_m  # [kg/m^2]

# porous material properties
d = 0.18  # [m] depth of porous material
Zp, kp = jca(f, sigma=sigma, tortuosity=tortuosity, porosity=porosity, viscous_characteristic_length=viscous_characteristic_length)  # JCA parameters

# estimating the ressonance frequency —— in reality, it will be somewhere between f50 and f60
membrane1_f50 = 50 / np.sqrt(surface_density * D)
membrane1_f60 = 60 / np.sqrt(surface_density * D)
Zm = 1j * omega * surface_density  # membrane impedance

# fmt: off
Zsp = -1j*Zp / np.tan(kp*d) # impedance on top of the porous material layer
Zsi = (((-1j*Zsp*rho0*c0) / (np.tan(k0*(D - d)))) + (rho0*c0)**2) / (Zsp - ((1j*rho0*c0) / (np.tan(k0*(D - d))))) # impedance on top of the air layer
# fmt: on

Zs = Zm + Zsi  # total impedance (series association)

Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))  # reflection coefficient
alpha_membrane1 = 1 - (abs(Vp)) ** 2  # absorption coefficient
```

```python
# membrane 2 properties
D = 0.25  # [m] size of the air gap
membrane_thickness = 2e-3  # [m] membrane thickness
rho_m = 2697  # [kg/m^3] specific mass of the membrane
surface_density = membrane_thickness * rho_m  # [kg/m^2]

# porous material properties
d = 0.18  # [m] depth of porous material
Zp, kp = jca(f, sigma=sigma, tortuosity=tortuosity, porosity=porosity, viscous_characteristic_length=viscous_characteristic_length)  # JCA parameters

# estimating the ressonance frequency —— in reality, it will be somewhere between f50 and f60
membrane2_f50 = 50 / np.sqrt(surface_density * D)
membrane2_f60 = 60 / np.sqrt(surface_density * D)
Zm = 1j * omega * surface_density  # membrane impedance

# fmt: off
Zsp = -1j*Zp / np.tan(kp*d) # impedance on top of the porous material layer
Zsi = (((-1j*Zsp*rho0*c0) / (np.tan(k0*(D - d)))) + (rho0*c0)**2) / (Zsp - ((1j*rho0*c0) / (np.tan(k0*(D - d))))) # impedance on top of the air layer
# fmt: on

Zs = Zm + Zsi  # total impedance (series association)

Vp = (Zs - (rho0 * c0)) / (Zs + (rho0 * c0))  # reflection coefficient
alpha_membrane2 = 1 - (abs(Vp)) ** 2  # absorption coefficient
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_title(r"Combinação de absorvedores")
ax.set_xlabel("Frequência [Hz]")
ax.set_ylabel(r"$\alpha$ [-]")
ax.set_xlim(f_start, f_end)
ax.set_ylim(-0.2, 1.2)
ax.grid(axis="both", which="both", linestyle="--", alpha=0.4)

ax.semilogx(f, alpha_porous, label="Poroso")
ax.semilogx(f, alpha_perforated, label="Perfurado")
ax.semilogx(f, alpha_membrane1, label="Membrana 1")
ax.vlines(x=membrane1_f50, ymin=0.96, ymax=2, linestyles="dotted", alpha=0.9, colors="green", linewidth=1.3)
ax.vlines(x=membrane1_f60, ymin=0.99, ymax=2, linestyles="dotted", alpha=0.9, colors="green", linewidth=1.3)
ax.semilogx(f, alpha_membrane2, label="Membrana 2")
ax.vlines(x=membrane2_f50, ymin=0.95, ymax=2, linestyles="dotted", alpha=0.9, colors="red", linewidth=1.3)
ax.vlines(x=membrane2_f60, ymin=0.9, ymax=2, linestyles="dotted", alpha=0.9, colors="red", linewidth=1.3)

ax.legend(loc="lower right")
plt.savefig("Apresentações/Etapa 2/imagens/combinacao_absorvedores.svg", format="svg", bbox_inches="tight")
plt.show()
```
